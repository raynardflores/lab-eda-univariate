```python
# Loading Data Set

import pandas as pd

import matplotlib.pyplot as plt

# Load the dataset
file_path = 'amz_uk_price_prediction_dataset.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Frequency Tables

# Generate a frequency table for the product category

category_freq = data['category'].value_counts()

# Identify the top 5 most listed product categories

top_5_categories = category_freq.head(5)
top_5_categories
```

```python
# Bar chart for distribution of products across different categories

plt.figure(figsize=(10, 6))
category_freq.plot(kind='bar')
plt.title('Distribution of Products Across Different Categories')
plt.xlabel('Product Category')
plt.ylabel('Number of Listings')
plt.show()
```

```python
# Pie chart for top 5 categories

plt.figure(figsize=(8, 8))
top_5_categories.plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 5 Product Categories by Listings')
plt.ylabel('')
plt.show()
```

```python
# Part 2 - Delving into Product Pricing
```

```python
# Measures of Centrality

# Calculate the mean, median, and mode for the price of products
mean_price = data['price'].mean()
median_price = data['price'].median()
mode_price = data['price'].mode()[0]

mean_price, median_price, mode_price
```

```python
# Measures of Dispersion

# Determine the variance, standard deviation, range, and interquartile range for product price
variance_price = data['price'].var()
std_dev_price = data['price'].std()
range_price = data['price'].max() - data['price'].min()
iqr_price = data['price'].quantile(0.75) - data['price'].quantile(0.25)

variance_price, std_dev_price, range_price, iqr_price
```

```python
# Visualizations

# Histogram for product prices
plt.figure(figsize=(10, 6))
plt.hist(data['price'], bins=50, edgecolor='k')
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Number of Products')
plt.show()

# Box plot for product prices
plt.figure(figsize=(10, 6))
plt.boxplot(data['price'], vert=False)
plt.title('Box Plot of Product Prices')
plt.xlabel('Price')
plt.show()
```

```python
# Part 3 - Unpacking Product Ratings
```

```python
# Measures of Centrality

# Calculate the mean, median, and mode for the rating of products
mean_rating = data['stars'].mean()
median_rating = data['stars'].median()
mode_rating = data['stars'].mode()[0]

mean_rating, median_rating, mode_rating

```

```python
# Measures of Dispersion

# Determine the variance, standard deviation, and interquartile range for product rating

variance_rating = data['stars'].var()
std_dev_rating = data['stars'].std()
iqr_rating = data['stars'].quantile(0.75) - data['stars'].quantile(0.25)

variance_rating, std_dev_rating, iqr_rating
```

```python
# Shape of the Distribution

from scipy.stats import skew, kurtosis

# Calculate the skewness and kurtosis for the rating column
skewness_rating = skew(data['stars'].dropna())
kurtosis_rating = kurtosis(data['stars'].dropna())

skewness_rating, kurtosis_rating
```

```python
# Visualizations

# Histogram for product ratings
plt.figure(figsize=(10, 6))
plt.hist(data['stars'], bins=10, edgecolor='k')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Products')
plt.show()
```
